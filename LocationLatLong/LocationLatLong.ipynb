#Extracting Latitude and Longitude of divisions and sub-divisons of South Asian Countries. 

Latitude and Longitude of [Souhth Asian](https://www.library.illinois.edu/ias/sacollection/sa_countries/) cities / districts / divisons / atolls / localities (whatever availble for the place) are extracted from [City Population](https://www.citypopulation.de/). The latitide and longitude of those areas is then found using [GeoPy](https://geopy.readthedocs.io/en/stable/). 

```
!pip install geopandas
!pip install geopy
```

```
import pandas as pd
from  geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='myapplication')
```

```
def find_lat_long(cityList,country):
  for city in cityList:
    loc = geolocator.geocode(city+','+ country)
    try:
      print(city+country.upper()+","+str(loc.latitude)+","+ str(loc.longitude))
    except AttributeError:
      pass
```

##Pakistan


```
df_pak_areas= pd.read_html("https://www.citypopulation.de/php/pakistan-distr-admin.php")
df_pak_areas_table = df_pak_areas[0]
```

```
pak_areas = df_pak_areas_table[(df_pak_areas_table['Status'] != "Republic") & (df_pak_areas_table['Status'] != "Adjoining Tribal Area")]['Name']
pak_areas = pak_areas.to_list()
pak_areas = set(pak_areas)
```

```
len(pak_areas)
```

```
find_lat_long(pak_areas,"Pakistan")
```

```
df_india_areas= pd.read_html("https://www.citypopulation.de/php/india-admin.php")
df_india_areas_table = df_india_areas[0]
india_areas = df_india_areas_table[(df_india_areas_table['Status'] != "Republic")]['Name']
india_areas = india_areas.to_list()
india_areas = set(india_areas)
len(india_areas)
```

## Sri Lanka
Cities

```
df_sri_cities= pd.read_html("https://www.citypopulation.de/en/srilanka/cities/")
df_sri_cities_table = df_sri_cities[2]
sri_cities = df_sri_cities_table['Name']
sri_cities = sri_cities.to_list()
sri_cities = set(sri_cities)
len(sri_cities)
```

Divisions

```
df_sri_divisons= pd.read_html("https://www.citypopulation.de/en/srilanka/admin/")
df_sri_divisons_table = df_sri_divisons[0]
sri_divisons = df_sri_divisons_table[(df_sri_divisons_table['Status'] != "Republic")]['Name']
sri_divisons = sri_divisons.to_list()
sri_divisons = set(sri_divisons)
len(sri_divisons)
```

```
sri_areas = sri_cities.union(sri_divisons)
len(sri_areas)
```

```
find_lat_long(sri_areas,"Srilanka")
```
